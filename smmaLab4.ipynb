```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/Exp 4 HashTag Dataset (1).csv")

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# 1. Hashtag Usage Over Time
hashtag_usage = df.groupby(['Date', 'Hashtag'])['Usage Count'].sum().unstack()

# 2. Reach and Engagement Analysis
reach_by_hashtag = df.groupby('Hashtag')['Reach'].sum().sort_values(ascending=False)
engagement_by_hashtag = df.groupby('Hashtag')['Engagements'].sum().sort_values(ascending=False)

# 3. Top-Performing Hashtags
top_hashtags = df.groupby('Hashtag')[['Reach', 'Engagements']].sum().sort_values(by='Engagements', ascending=False).head(5)
print("Top 5 Performing Hashtags:\n", top_hashtags)

# 4. Platform Analysis
platform_performance = df.groupby('Platform')[['Reach', 'Engagements']].sum()
print("Performance by Platform:\n", platform_performance)

# Visualization
plt.figure(figsize=(14, 10))

# Hashtag Usage Over Time
plt.subplot(2, 2, 1)
hashtag_usage.plot(ax=plt.gca())
plt.title("Hashtag Usage Over Time")
plt.ylabel("Usage Count")
plt.xticks(rotation=45)

# Reach by Hashtag
plt.subplot(2, 2, 2)
sns.barplot(x=reach_by_hashtag.index, y=reach_by_hashtag.values, palette='viridis')
plt.title("Total Reach by Hashtag")
plt.xticks(rotation=45)
plt.ylabel("Reach")

# Engagement by Hashtag
plt.subplot(2, 2, 3)
sns.barplot(x=engagement_by_hashtag.index, y=engagement_by_hashtag.values, palette='magma')
plt.title("Total Engagement by Hashtag")
plt.xticks(rotation=45)
plt.ylabel("Engagements")

# Platform Performance
plt.subplot(2, 2, 4)
sns.barplot(x=platform_performance.index, y=platform_performance['Engagements'], palette='coolwarm')
plt.title("Engagements by Platform")
plt.xticks(rotation=45)
plt.ylabel("Engagements")

plt.tight_layout()
plt.show()
```
