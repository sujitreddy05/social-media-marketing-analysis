```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulated influencer data (replace this with real API data)
data = {
    "post_id": [1, 2, 3, 4, 5],
    "post_date": ["2024-11-01", "2024-11-06", "2024-11-12", "2024-11-18", "2024-11-22"],
    "likes": [2200, 2500, 2100, 2700, 2400],
    "comments": [80, 110, 95, 120, 105],
    "shares": [30, 45, 50, 60, 40],  # Added shares metric
    "followers_gained": [180, 230, 210, 320, 280],
    "authentic_engagement_rate": [6.2, 6.5, 6.0, 7.0, 6.8]  # Authentic engagement rate
}

# Convert to DataFrame
df = pd.DataFrame(data)



# Analyze the data
df['post_date'] = pd.to_datetime(df['post_date'])
df['engagement_rate'] = (df['likes'] + df['comments']) / df['followers_gained']

# Plot the results
plt.figure(figsize=(14, 7))

# Plot engagement rate over time
plt.subplot(2, 1, 1)
plt.plot(df['post_date'], df['engagement_rate'], marker='o', linestyle='-', color='b')
plt.title('Engagement Rate Over Time')
plt.xlabel('Post Date')
plt.ylabel('Engagement Rate')
plt.grid(True)

# Plot authentic engagement rate
plt.subplot(2, 1, 2)
plt.plot(df['post_date'], df['authentic_engagement_rate'], marker='o', linestyle='-', color='r')
plt.title('Authentic Engagement Rate Over Time')
plt.xlabel('Post Date')
plt.ylabel('Authentic Engagement Rate (%)')
plt.grid(True)

plt.tight_layout()
plt.show()

# Calculate overall metrics
average_engagement_rate = df['engagement_rate'].mean()
average_authentic_engagement_rate = df['authentic_engagement_rate'].mean()

# Print the results
print(f"Average Engagement Rate: {average_engagement_rate:.2f}")
print(f"Average Authentic Engagement Rate: {average_authentic_engagement_rate:.2f}%")

# Recommendation
if average_authentic_engagement_rate > 4.0:  # Example threshold
    print("This influencer has high authentic engagement and would be a good fit for the campaign.")
else:
    print("This influencer may not be a good fit due to low authentic engagement.")
```
