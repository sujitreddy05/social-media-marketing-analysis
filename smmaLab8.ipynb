```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/Exp 8 Dataset(1).csv")

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# 1. Touchpoint Distribution
touchpoint_counts = df['Touchpoint'].value_counts()

# 2. Platform Engagement
platform_engagement = df.groupby('Platform')['Engagement Score'].mean().sort_values(ascending=False)

# 3. Conversion Rate by Platform
conversion_by_platform = df.groupby('Platform')['Conversion'].mean() * 100

# 4. Session Duration Analysis
session_duration_by_touchpoint = df.groupby('Touchpoint')['Session Duration'].mean().sort_values()

# 5. Device Analysis
device_usage = df['Device'].value_counts()

# Visualization
plt.figure(figsize=(14, 10))

# Touchpoint Distribution
plt.subplot(2, 2, 1)
sns.barplot(x=touchpoint_counts.index, y=touchpoint_counts.values, palette='coolwarm')
plt.title("Touchpoint Distribution")
plt.xticks(rotation=45)
plt.ylabel("Count")

# Platform Engagement
plt.subplot(2, 2, 2)
sns.barplot(x=platform_engagement.index, y=platform_engagement.values, palette='viridis')
plt.title("Average Engagement Score by Platform")
plt.xticks(rotation=45)
plt.ylabel("Engagement Score")

# Conversion Rate by Platform
plt.subplot(2, 2, 3)
sns.barplot(x=conversion_by_platform.index, y=conversion_by_platform.values, palette='magma')
plt.title("Conversion Rate by Platform (%)")
plt.xticks(rotation=45)
plt.ylabel("Conversion Rate (%)")

# Device Usage
plt.subplot(2, 2, 4)
sns.barplot(x=device_usage.index, y=device_usage.values, palette='coolwarm')
plt.title("Device Usage Distribution")
plt.xticks(rotation=45)
plt.ylabel("Count")

plt.tight_layout()
plt.show()

# Print insights
print("Session Duration by Touchpoint:\n", session_duration_by_touchpoint)
```
